<h2>Binary image</h2>

```python
img = cv2.imread("../src/task_2s/images/bridge_path_color.png")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB

plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(grey, (5, 5), 0)

plt.imshow(blurred, cmap='gray')
plt.axis('off')
plt.show()
```

```python
_, binary = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.imshow(binary, cmap='gray')
plt.axis('off')
plt.show()
```

```python
closed = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, np.ones((20, 20), np.uint8))
plt.imshow(closed, cmap='gray')
plt.axis('off')
plt.show()
```

```python
eroded = cv2.erode(closed, np.ones((20, 20), np.uint8), iterations=1)

plt.imshow(eroded, cmap='gray')
plt.axis('off')
plt.show()
```

```python
dilated = cv2.dilate(eroded, np.ones((10, 10), np.uint8), iterations=1)
plt.imshow(dilated, cmap='gray')
plt.axis('off')
plt.show()
```
